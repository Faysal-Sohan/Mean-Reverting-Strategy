```python
import pandas as pd
import yfinance as yf
```

```python

# Download S&P 500 ticker symbols
snp500_url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
data = pd.read_html(snp500_url)
sp500_tickers = data[0]['Symbol'].tolist()
# Correcting ticker name for Berkshire 
i = sp500_tickers.index("BRK.B")
sp500_tickers[i] = "BRK-B"
# Correcting ticker name for Brown Foreman 
i = sp500_tickers.index("BF.B")
sp500_tickers[i] = "BF-B"
```

```python
# Defining Start & End Date
start_date = "2010-01-01"
end_date = "2024-01-01"
for ticker in sp500_tickers:
    try:
        data = yf.download(ticker, start=start_date, end=end_date, progress=False)
        data.to_csv(f"../data/S & P 500 Assets/daily/{ticker}.csv")
        # processed_data = csv_data_process(data)
        # processed_data.to_csv(f"/home/sohan/Documents/csv commodities/S & P 500 Assets/{ticker}.csv")
        # processed_data.to_csv(f"/home/sohan/Documents/csv commodities/S & P 100/daily/{ticker}.csv", index=False)
        print(f"Downloaded data for {ticker}")
    except Exception as e:
        print(f"Error downloading data for {ticker}: {e}")
```

```python
def csv_data_process(df: pd.DataFrame):
    df.drop("Adj Close", axis=1,inplace=True)
    new_names = {'Date': 'date', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close', 'Volume': 'volume'}
    df.rename(columns=new_names, inplace=True)
    df.iloc[:,1:] = df.iloc[:,1:].apply(lambda x: x.astype(float))
    df['timestamp'] = pd.to_datetime(df['date'])
    df.drop('date', axis=1, inplace=True)
    df = df.reindex(columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df['timestamp'] = df['timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S.%f')
    return df
```

```python
for ticker in sp500_tickers:
    try:
        data = pd.read_csv(f"../data/S & P 500 Assets/daily/{ticker}.csv")
        processed_data = csv_data_process(data)
        processed_data.to_csv(f"/home/sohan/Documents/csv commodities/S & P 500 Assets/daily/{ticker}.csv",index=False)
    except Exception as e:
        print(f"Error Occured for {ticker}. {e}")
```
